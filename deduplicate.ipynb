```python
import pandas as pd
import numpy as np
from tqdm import tqdm
from fuzzywuzzy import fuzz
tqdm.pandas()
```

```python
df = pd.read_parquet('data/ar-en-final.parquet')
```

```python
df.head(10)
```

```python
def get_similarity_ratio(str1, str2):
    return fuzz.ratio(str1, str2)
```

```python
get_similarity_ratio(*list(df.loc[9][['en','ar']]))
```

```python
df['similarity_ratio'] = df.progress_apply(lambda x: get_similarity_ratio(x['en'], x['ar']), axis=1)
```

```python
df_deduplicated = df[df['similarity_ratio'] <= 75].drop('similarity_ratio', axis=1)
```

```python
df_deduplicated
```

```python
df_deduplicated.to_parquet('data/ar-en-final-fuzzy-deduplicated.parquet')
```
